```python
from math import inf


def ed(X,Y):
    n = len(X)
    m = len(Y)
    T = [ [inf] * (m+1) for _ in range(n+1)]
    T[0][0] = 0
    for i in range(n+1):
        for j in range(m+1):
            if i - 1 == j:
                continue
            if i > 0:
                T[i][j] = min(T[i][j],T[i-1][j]+1)
            if j > 0:
                T[i][j] = min(T[i][j],T[i][j-1]+1)
            if i > 0 and j > 0:
                if X[i-1] == Y[j-1]:
                    T[i][j] = min(T[i][j],T[i-1][j-1])
                else:
                    T[i][j] = min(T[i][j],T[i-1][j-1]+1)
    return T

def self_ed(X):
    n = len(X)
    T = [ [inf] * (n+1) for _ in range(n+1)]
    pr = [[(-1, -1)] * (n + 1) for _ in range(n + 1)]
    T[0][0] = 0
    for i in range(n+1):
        for j in range(n+1):
            if i > 0 and (T[i][j] > T[i - 1][j] + 1 or (T[i][j] == T[i - 1][j] + 1 and i - 1 == j)):
                T[i][j] = T[i-1][j] + 1
                pr[i][j] = (i - 1, j)
            if j > 0 and (T[i][j] > T[i][j - 1] + 1 or (T[i][j] == T[i][j - 1] + 1 and i == j - 1)):
                T[i][j] = T[i][j-1] + 1
                pr[i][j] = (i, j - 1)
            if i > 0 and j > 0 and i != j:
                if X[i-1] == X[j-1] and T[i][j] > T[i - 1][j - 1]:
                    T[i][j] = T[i-1][j-1]
                    pr[i][j] = (i - 1, j - 1)
                elif T[i][j] > T[i - 1][j - 1] + 1:
                    T[i][j] = T[i-1][j-1] + 1
                    pr[i][j] = (i - 1, j - 1)
    return T, pr
```

```python
X = 'aaabcbcbccc'
T, pr = self_ed(X)
path = []
cur = (len(X), len(X))
cnt_on = 0
while cur[0] != -1:
    path.append(cur)
    cur = pr[cur[0]][cur[1]]
path = list(reversed(path))
print(path)
print(T)
```
